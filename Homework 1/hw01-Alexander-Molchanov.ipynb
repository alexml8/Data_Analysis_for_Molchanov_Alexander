- –ù–∞–π–¥–∏—Ç–µ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –ø–æ–¥–º–∞—Ç—Ä–∏—Ü—ã –º–∞—Ç—Ä–∏—Ü—ã $X$, —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–Ω–æ–π –Ω–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–∏ 2 –∏ 3 —Å—Ç—Ä–æ–∫–∏ –∏ 1 –∏ 2 —Å—Ç–æ–ª–±—Ü–∞; —Å—á–∏—Ç–∞–µ–º, —á—Ç–æ —Å—Ç—Ä–æ–∫–∏ –∏ —Å—Ç–æ–ª–±—Ü—ã –Ω—É–º–µ—Ä—É—é—Ç—Å—è —Å –µ–¥–∏–Ω–∏—Ü—ã (–∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ slicing!). –¢–∞–∫–æ–π –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –Ω–∞–∑—ã–≤–∞–µ—Ç—Å—è **–º–∏–Ω–æ—Ä–æ–º** –º–∞—Ç—Ä–∏—Ü—ã $X$.
- –ù–∞–π–¥–∏—Ç–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ $X^TX$, –µ–≥–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –∏ —Å–ª–µ–¥.

```python
# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã 4x5
X = np.random.rand(4, 5)
print(f"–°–ª—É—á–∞–π–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞ X:\n{X}\n")

# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –º–∏–Ω–æ—Ä–∞ –º–∞—Ç—Ä–∏—Ü—ã X
minor = np.linalg.det(X[1:3, 0:2])
print(f"–ú–∏–Ω–æ—Ä –º–∞—Ç—Ä–∏—Ü—ã X:\n{minor}\n")

# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è X^T * X
XTX = np.dot(X.T, X)
print(f"–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ X^T * X:\n{XTX}\n")

# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è X^T * X
determinant_XTX = np.linalg.det(XTX)
print(f"–û–ø—Ä–µ–¥–µ–ª–∏—Ç–µ–ª—å –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è X^T * X:\n{determinant_XTX}\n")

# –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ —Å–ª–µ–¥–∞ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è X^T * X
tr_XTX = np.trace(XTX)
print(f"–°–ª–µ–¥ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è X^T * X:\n{tr_XTX}\n")
```

–ù–∞ –ø–∞—Ä–µ –º—ã –≥–æ–≤–æ—Ä–∏–ª–∏, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Ü–∏–∫–ª–æ–≤ (–∏ —Å–∞–º–æ–ø–∏—Å–Ω—ã—Ö –≤—ã—á–∏—Å–ª–µ–Ω–∏–π –≤ —Ü–µ–ª–æ–º) –≤ python –ª—É—á—à–µ –ø–æ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –∏–∑–±–µ–≥–∞—Ç—å –∏ –≤–∞–∂–Ω–æ —É–º–µ—Ç—å –Ω–∞—Ö–æ–¥–∏—Ç—å —Å–ø–æ—Å–æ–±—ã –¥–µ–ª–∞—Ç—å –≤—Å—ë –±–∏–±–ª–∏–æ—Ç–µ—á–Ω—ã–º–∏ —Å—Ä–µ–¥—Å—Ç–≤–∞–º–∏.

–í —Å–ª–µ–¥—É—é—â–∏—Ö —Ç—Ä—ë—Ö –∑–∞–¥–∞–Ω–∏—è—Ö –Ω–µ–ª—å–∑—è –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è —Ü–∏–∫–ª–∞–º–∏ –≤–æ–æ–±—â–µ, –∞ —Ç–∞–∫–∂–µ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏—è–º–∏ `map` –∏ `reduce` –∏ –∏–º –ø–æ–¥–æ–±–Ω—ã–º–∏; –≤–º–µ—Å—Ç–æ —ç—Ç–æ–≥–æ –ø–æ—Å—Ç–∞—Ä–∞–π—Ç–µ—Å—å —Å–≤–µ—Å—Ç–∏ –≤—Å—ë –∫ –º–∞—Ç—Ä–∏—á–Ω—ã–º –æ–ø–µ—Ä–∞—Ü–∏—è–º –∏–∑ `numpy` (–Ω–æ, –æ–ø—è—Ç—å –∂–µ, –Ω–µ `np.vectorize` –∏–ª–∏ —á–µ–º—É-—Ç–æ –ø–æ–¥–æ–±–Ω–æ–º—É). –ß—Ç–æ–±—ã —É–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ –ø–æ–ª—É—á–∏–ª–æ—Å—å –∏–º–µ–Ω–Ω–æ —Ç–æ, —á—Ç–æ –Ω—É–∂–Ω–æ, –Ω–∞–ø–∏—à–∏—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ —Ç–µ—Å—Ç—ã —Å–æ —Å–ª—É—á–∞–π–Ω—ã–º–∏ –º–∞—Ç—Ä–∏—Ü–∞–º–∏.

**[5 –±–∞–ª–ª–æ–≤]** –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `prod_and_sq_sum(A)`, –≤—ã—á–∏—Å–ª—è—é—â—É—é –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∏ —Å—É–º–º—É –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–π –º–∞—Ç—Ä–∏—Ü—ã `A`.

```python
def prod_and_sq_sum(A):
    el = np.diagonal(A)
    prod = np.prod(el)
    sum_of_squares = np.sum(el ** 2)

    return prod, sum_of_squares
```

**[5 –±–∞–ª–ª–æ–≤]** –î–ª—è –º–∞—Ç—Ä–∏—Ü `A` –∏ `B` —Ä–∞–∑–º–µ—Ä–∞ $m\times n$ –æ–±–æ–∑–Ω–∞—á–∏–º —á–µ—Ä–µ–∑ $a_1,\ldots,a_n$ –∏ $b_1,\ldots,b_n$ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –∏—Ö —Å—Ç–æ–ª–±—Ü—ã. –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `f(A, B, k)`, –≤—ã—á–∏—Å–ª—è—é—â—É—é

$$\sum_{i=1}^{\min(k, n)}a_ib_i^T$$

```python
def f(A, B, k):
    n = min(k, A.shape[1])
    result = np.sum(np.outer(A[:, :n], B[:, :n]), axis = 2)
    return result
```

**[5 –±–∞–ª–ª–æ–≤]** –ù–∞–ø–∏—à–∏—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é `get_diag(A,B)`, –ø—Ä–∏–Ω–∏–º–∞—é—â—É—é –¥–≤–µ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã—Ö –º–∞—Ç—Ä–∏—Ü—ã `A` –∏ `B` –æ–¥–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –∏ –≤–æ–∑–≤—Ä–∞—â–∞—é—â—É—é –≤–µ–∫—Ç–æ—Ä –¥–∏–∞–≥–æ–Ω–∞–ª—å–Ω—ã—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏—è `AB`, –Ω–µ –≤—ã—á–∏—Å–ª—è—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ü–µ–ª–∏–∫–æ–º. 

```python
def get_diag(A, B):
    assert A.shape == B.shape, "–ú–∞—Ç—Ä–∏—Ü—ã A –∏ B —Ä–∞–∑–Ω—ã–µ, –∞ –¥–æ–ª–∂–Ω—ã –∏–º–µ—Ç—å –æ–¥–∏–Ω–∞–∫–æ–≤—ã–π —Ä–∞–∑–º–µ—Ä"
    diagonal = np.einsum('ij,ji->i', A, B)

    return diagonal
```

## –ó–∞–¥–∞—á–∞ 2: –ø—Ä–æ —É–¥–∞—á—É –∏ –Ω–∞—à—É –∂–∏–∑–Ω—å (25 –±–∞–ª–ª–æ–≤)

–ó–∞–¥–∞—á–∞ –±—ã–ª–∞ –≤–¥–æ—Ö–Ω–æ–≤–ª–µ–Ω–∞ [–≤–æ—Ç —ç—Ç–∏–º –≤–∏–¥–µ–æ.](https://www.youtube.com/watch?v=3LopI4YeC4I)

–õ—é–¥–∏ —á–∞—Å—Ç–æ –Ω–µ–¥–æ–æ—Ü–µ–Ω–∏–≤–∞—é—Ç —Ä–æ–ª—å —Å–ª—É—á–∞–π–Ω–æ—Å—Ç–∏ –≤ —Å–≤–æ–µ–π –∂–∏–∑–Ω–∏. –ò–º –∫–∞–∂–µ—Ç—Å—è, —á—Ç–æ –µ—Å–ª–∏ –æ–Ω–∏ –ø—Ä–æ–∫–∞—á–∞—é—Ç —Å–≤–æ–∏ –Ω–∞–≤—ã–∫–∏ –∏ –æ—Ç–¥–∞–¥—É—Ç—Å—è –ø–æ–ª–Ω–æ—Å—Ç—å—é —Ä–∞–±–æ—Ç–µ, –æ–Ω–∏ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç —É—Å–ø–µ—Ö–∞. –õ—é–¥–∏ –∑–∞–±—ã–≤–∞—é—Ç, —á—Ç–æ —É–¥–∞—á–∞ —Ç–æ–∂–µ –≤–∞–∂–Ω–∞. –ò–Ω–æ–≥–¥–∞ —Å–ª–∏—à–∫–æ–º.

–ü—É—Å—Ç—å —É –Ω–∞—Å –≤ –≤—ã–±–æ—Ä–∫–µ –µ—Å—Ç—å $10000$ –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã. –ò–∑ –Ω–∏—Ö —Ç–æ–ª—å–∫–æ $10$ —á–µ–ª–æ–≤–µ–∫ –ø—Ä–æ–π–¥—É—Ç –≤—Å–µ –∏—Å–ø—ã—Ç–∞–Ω–∏—è –∏ –ø—Ä–∏—Å—Ç—É–ø—è—Ç –∫ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–∫–∞–º. –û—Ç–±–æ—Ä –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã –Ω–∞ $95\%$ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç –Ω–∞–≤—ã–∫–æ–≤ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –∏ –Ω–∞ $5\%$ –æ—Ç –µ–≥–æ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç–∏. –ù–∞–≤—ã–∫–∏ –∏ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç—å –Ω–µ –∑–∞–≤–∏—Å—è—Ç –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞.

__–∞) [5 –±–∞–ª–ª–æ–≤]__  –°–≥–µ–Ω–µ—Ä–∏—Ä—É–π—Ç–µ –Ω–∞–≤—ã–∫ –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞ –∏–∑ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[0; 1]$. –ü–æ –∞–Ω–∞–ª–æ–≥–∏–∏ —Å–≥–µ–Ω–∏—Ä–∏—Ä—É–π—Ç–µ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ –∫–∞–Ω–¥–∏–¥–∞—Ç–∞. –ò—Ç–æ–≥–æ–≤—É—é –≤–µ–ª–∏—á–∏–Ω—É, –ø–æ –∫–æ—Ç–æ—Ä–æ–π –±—É–¥–µ—Ç –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—å—Å—è –æ—Ç–±–æ—Ä –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã, —Ä–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ –ø–æ —Ñ–æ—Ä–º—É–ª–µ: 

$$
\text{score} = 0.95 \cdot \text{skill} + 0.05 \cdot \text{luck}.
$$

–ü–æ–≤—Ç–æ—Ä–∏—Ç–µ —ç—Ç—É —Å–∏–º—É–ª—è—Ü–∏—é $10^4$ —Ä–∞–∑. –û—Ü–µ–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç–∏ —Ç–µ—Ö, –∫—Ç–æ —Å—Ç–∞–ª –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —É–¥–∞—á–∏ —Å—Ä–µ–¥–∏ —Ç–µ—Ö, –∫—Ç–æ —Å—Ç–∞–ª –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º.

**Hint:** –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é `np.argpartition`. –û–Ω–∞ –ø–æ–º–æ–∂–µ—Ç –≤–∞–º —Å–¥–µ–ª–∞—Ç—å –∫–æ–¥ –ø—Ä–æ—â–µ. –¢–∞–∫–∂–µ –æ–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ –≤ –ø—É–Ω–∫—Ç–∞—Ö –±) –∏ –≤) –Ω—É–∂–Ω–æ –ø–µ—Ä–µ–∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–∏–º—É–ª—è—Ü–∏–π –∏–∑ –ø—É–Ω–∫—Ç–∞ –∞). –ò–Ω–∞—á–µ –∫–æ–¥ –±—É–¥–µ—Ç —Ä–∞–±–æ—Ç–∞—Ç—å —Å–ª–∏—à–∫–æ–º –¥–æ–ª–≥–æ.

```python
n_simulations = 10000
n_candidates = 10000
n_astronauts = 10
```

```python
avg_luck = np.zeros(n_simulations)
avg_intersection = np.zeros(n_simulations)
n_range = np.arange(10, 501, 10)
perc_intersection = np.zeros(len(n_range))

for i in range(n_simulations):
    skills = np.random.uniform(0, 1, n_candidates)
    luck = np.random.uniform(0, 1, n_candidates)
    scores = 0.95 * skills + 0.05 * luck

    top_by_skills = np.argpartition(skills, -n_astronauts)[-n_astronauts:]
    top_by_scores = np.argpartition(scores, -n_astronauts)[-n_astronauts:]

    intersection = np.intersect1d(top_by_skills, top_by_scores).size
    avg_intersection[i] = intersection

    avg_luck[i] = luck[top_by_scores].mean()

    if i == 0:
        for j, n in enumerate(n_range):
            top_by_skills_n = np.argpartition(skills, -n)[-n:]
            top_by_scores_n = np.argpartition(scores, -n)[-n:]
            intersection_n = np.intersect1d(top_by_skills_n, top_by_scores_n).size
            perc_intersection[j] = (intersection_n / n) * 100

mean_luck = avg_luck.mean()
mean_intersection = avg_intersection.mean() * 10
```

```python
plt.figure(figsize = (10, 6))
plt.hist(avg_luck, bins = 30, alpha = 0.75, color = 'orange')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É–¥–∞—á–ª–∏–≤–æ—Å—Ç–∏ —Å—Ä–µ–¥–∏ –≤—ã–±—Ä–∞–Ω–Ω—ã—Ö –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤')
plt.xlabel('–£–¥–∞—á–ª–∏–≤–æ—Å—Ç—å')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.grid(True)
plt.show()

print(f"–°—Ä–µ–¥–Ω—è—è —É–¥–∞—á–∞ –∑–∞—á–∏—Å–ª–µ–Ω–∏—è: {mean_luck:.3f}")
```

__–±) [5 –±–∞–ª–ª–æ–≤]__  –û—Ç–±–µ—Ä–∏—Ç–µ –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤ –ø–æ –Ω–∞–≤—ã–∫–∞–º, `skill`. –û—Ü–µ–Ω–∏—Ç–µ —Å –ø–æ–º–æ—â—å—é —Å–∏–º—É–ª—è—Ü–∏–π, –Ω–∞ —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–ª–æ–≤–µ–∫ —Ç–æ–ø –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –ø–æ `skill` –ø–µ—Ä–µ—Å–µ–∫–∞–µ—Ç—Å—è —Å —Ç–æ–ø–æ–º –ø–æ `score`.

```python
plt.figure(figsize = (10, 6))
plt.hist(avg_intersection * 10, bins = np.arange(0, 100, 10), alpha = 0.75, color = 'orange')
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ—Ü–µ–Ω—Ç–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è —Ç–æ–ø-–∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –ø–æ –Ω–∞–≤—ã–∫–∞–º –∏ –æ–±—â–µ–º—É –±–∞–ª–ª—É')
plt.xlabel('–ü—Ä–æ—Ü–µ–Ω—Ç –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.xticks(np.arange(0, 100, 10))
plt.grid(True)
plt.show()

print(f"–í –≤ —Å—Ä–µ–¥–Ω–µ–º {mean_intersection:.3f}% –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è")
```

__–≤) [5 –±–∞–ª–ª–æ–≤]__ –ê —á—Ç–æ –±—É–¥–µ—Ç, –µ—Å–ª–∏ –º—ã –±—É–¥–µ–º –æ—Ç–±–∏—Ä–∞—Ç—å –±–æ–ª—å—à–µ –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–≤? 

–î–ª—è –∫–∞–∂–¥–æ–≥–æ $n$ –æ—Ç $10$ –¥–æ $500$ —Å —à–∞–≥–æ–º –≤ $10$ —Å–¥–µ–ª–∞–π—Ç–µ –ø—Ä–æ—Ü–µ–¥—É—Ä—É –æ—Ç–±–æ—Ä–∞ –≤ –∫–æ—Å–º–æ–Ω–∞–≤—Ç—ã —Ç–æ–ø-$n$ –∫–∞–Ω–¥–∏–¥–∞—Ç–æ–≤ –ø–æ `skill` –∏ –ø–æ `score`. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ, –Ω–∞ —Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –ø—Ä–æ—Ü–µ–Ω—Ç–æ–≤ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è —ç—Ç–∏ —Ç–æ–ø—ã. –ò–∑–æ–±—Ä–∞–∑–∏—Ç–µ –ø–æ–ª—É—á–∏–≤—à—É—é—Å—å—è –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ. –ü–æ –æ—Å–∏ –∞–±—Å—Ü–∏—Å—Å –æ—Ç–ª–æ–∂–∏—Ç–µ $n$, –ø–æ –æ—Å–∏ –æ—Ä–¥–∏–Ω–∞—Ç –ø—Ä–æ—Ü–µ–Ω—Ç –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è.

__–≥) [5 –±–∞–ª–ª–æ–≤]__ –°–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥—ã –∏–∑ –ø—Ä–æ–¥–µ–ª–∞–Ω–Ω—ã—Ö —Ä–∞—Å—Å—á—ë—Ç–æ–≤.

–ü–æ –≥—Ä–∞—Ñ–∏–∫–∞–º –≤–∏–¥–Ω–æ, —á—Ç–æ –∫–æ–≥–¥–∞ –∏–º–µ–µ—Ç—Å—è –Ω–µ–±–æ–ª—å—à–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–∞–Ω–Ω—ã—Ö –≤ –≤—ã–±–æ—Ä–∫–µ, –≤–ª–∏—è–Ω–∏–µ —É–¥–∞—á–∏ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –≤—ã—à–µ, —á–µ–º –ø—Ä–∏ –±–æ–ª—å—à–µ–º –æ–±—ä–µ–º–µ –¥–∞–Ω–Ω—ã—Ö.

> –ü–æ–≤–æ–¥ –ª–∏ —ç—Ç–æ –≥—Ä—É—Å—Ç–∏—Ç—å? –ö–æ–Ω–µ—á–Ω–æ –∂–µ, –Ω–µ—Ç. –ï—Å–ª–∏ –≤—ã –ø–æ–ø—ã—Ç–∞–ª–∏—Å—å —Å—Ö–≤–∞—Ç–∏—Ç—å—Å—è –∑–∞ –æ–¥–Ω—É –∏–∑ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–µ–π –∏ —É –≤–∞—Å –Ω–∏—á–µ–≥–æ –Ω–µ –≤—ã—à–ª–æ, –∑–∞–±—É–¥—å—Ç–µ –æ–± —ç—Ç–æ–º. –ù–µ –≤–ø–∞–¥–∞–π—Ç–µ –≤ –¥–µ–ø—Ä–µ—Å—Å–∏—é, –ø—Ä–æ–¥–æ–ª–∂–∞–π—Ç–µ –ø—ã—Ç–∞—Ç—å—Å—è. –í –º–∏—Ä–µ –µ—Å—Ç—å —Å—Ç–æ–ª—å–∫–æ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –º–æ–∂–Ω–æ —É–ø—É—Å—Ç–∏—Ç—å. –û–±–∏–¥–Ω–æ –±—É–¥–µ—Ç, –µ—Å–ª–∏ –≤—ã –Ω–µ –ø–æ–ø—Ä–æ–±—É–µ—Ç–µ –µ—â—ë. –†–∞–Ω–æ –∏–ª–∏ –ø–æ–∑–¥–Ω–æ —É–¥–∞—á–∞ –≤–∞–º —É–ª—ã–±–Ω—ë—Ç—Å—è ‚ù§Ô∏è‚ù§Ô∏è‚ù§Ô∏è

–î–∞–≤–∞–π—Ç–µ –ø–æ–ø—Ä–æ–±—É–µ–º –ø–æ–Ω—è—Ç—å, –∫–æ–≥–¥–∞ –∏–º–µ–Ω–Ω–æ –æ–Ω–∞ –≤–∞–º —É–ª—ã–±–Ω—ë—Ç—Å—è. 

__–¥) [4 –±–∞–ª–ª–∞]__ –Æ—Ä–∞ —Ö–æ—á–µ—Ç —Å—Ç–∞—Ç—å –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º. –£ –Ω–µ–≥–æ –¥–æ–≤–æ–ª—å–Ω–æ —Ö–æ—Ä–æ—à–æ —Ä–∞–∑–≤–∏—Ç—ã –Ω–∞–≤—ã–∫–∏. –Æ—Ä–∞ –≤—Ö–æ–¥–∏—Ç –≤ $1\%$ –ª—É—á—à–∏—Ö, –µ–≥–æ `score = 0.99` (–Æ—Ä–∞ –ø—Ä–æ–∫–∞—á–∞–ª —Å–≤–æ–∏ –Ω–∞–≤—ã–∫–∏ –¥–æ 0.99 –∏ –≤–æ–æ–±—â–µ –Ω–µ –ø–æ–ª–∞–≥–∞–µ—Ç—Å—è –Ω–∞ —É–¥–∞—á—É). –° –∫–∞–∫–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –Æ—Ä–∞ —Å—Ç–∞–Ω–µ—Ç –∫–æ—Å–º–æ–Ω–∞–≤—Ç–æ–º?

```python
Yura_score = 0.99
Yura_selected = 0

for _ in range(n_simulations):
    skills = np.random.uniform(0, 1, n_candidates - 1)
    luck = np.random.uniform(0, 1, n_candidates - 1)
    scores = 0.95 * skills + 0.05 * luck

    all_scores = np.append(scores, Yura_score)

    if Yura_score >= np.partition(all_scores, -n_astronauts)[-n_astronauts]:
        Yura_selected += 1

Yura_probability = Yura_selected / n_simulations
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –Æ—Ä–∞ –ø–æ—Å—Ç—É–ø–∏—Ç: {Yura_probability}")
```

__–µ) [1 –±–∞–ª–ª]__ –Æ—Ä–∞ –ø—ã—Ç–∞–µ—Ç—Å—è —Å–Ω–æ–≤–∞ –∏ —Å–Ω–æ–≤–∞. –°–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –ø–æ–ø—ã—Ç–æ–∫ –µ–º—É –Ω–∞–¥–æ —Å–¥–µ–ª–∞—Ç—å –¥–æ –ø–µ—Ä–≤–æ–≥–æ —É—Å–ø–µ—Ö–∞? 

__Hint:__ —Ç—É—Ç –Ω–µ –Ω—É–∂–Ω–æ –Ω–∏—á–µ–≥–æ —Å–∏–º—É–ª–∏—Ä–æ–≤–∞—Ç—å. –ù–∞–¥–æ –≤—Å–ø–æ–º–Ω–∏—Ç—å –ø—Ä–æ –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏ –µ–≥–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ.

```python
average_attempts = 1 / Yura_probability
print(f"–Æ—Ä–∞ —Ç–æ—á–Ω–æ –ø–æ—Å—Ç—É–ø–∏—Ç, –µ—Å–ª–∏ –ø–æ–ø—Ä–æ–±—É–µ—Ç {round(average_attempts, 0):.0f} —Ä–∞–∑")
```

## –ó–∞–¥–∞—á–∞ 3: –ø–∞—Ä–∞–¥–æ–∫—Å –ª–∏—Ñ—Ç–∞ (20 –±–∞–ª–ª–æ–≤)

[–†–∞–π–∞–Ω –ì–æ—Å–ª–∏–Ω–≥](https://www.youtube.com/watch?v=2jT71iFvpSM), [–°—Ç–∏–≤ –†–æ–¥–∂–µ—Ä—Å](https://www.youtube.com/watch?v=hLUdF8cjzyA) –∏ [–≥—Ä—É–ø–ø–∞ Radiohead](https://www.youtube.com/watch?v=QBGaO89cBMI) –∫–∞—Ç–∞—é—Ç—Å—è –Ω–∞ –ª–∏—Ñ—Ç–µ.

–ü—Ä–æ–≤–µ–¥–∏—Ç–µ $10^4$ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤ –∏ —Å –∏—Ö –ø–æ–º–æ—â—å—é –æ—Ç–≤–µ—Ç—å—Ç–µ –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –≤–æ–ø—Ä–æ—Å—ã: 

__–∞) [5 –±–∞–ª–ª–æ–≤]__   –†–∞–π–∞–Ω –ì–æ—Å–ª–∏–Ω–≥ –≤ —Å–ª—É—á–∞–π–Ω–æ–µ –≤—Ä–µ–º—è –ø–æ–¥—Ö–æ–¥–∏—Ç –∫ –ª–∏—Ñ—Ç—É –Ω–∞ 3 —ç—Ç–∞–∂–µ –∏ –∂–¥—ë—Ç –µ–≥–æ. –û—Ü–µ–Ω–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ª–∏—Ñ—Ç –ø—Ä–∏–µ–¥–µ—Ç –∫ –†–∞–π–∞–Ω—É —Å–≤–µ—Ä—Ö—É. 

```python
n_simulations = 10000
```

```python
ryan_arrival_times = np.random.randint(0, 60, n_simulations)

lift_up_time = 3
lift_down_time = 57

lift_coming_from_above = np.sum((ryan_arrival_times >= lift_down_time) | (ryan_arrival_times < lift_up_time))

probability_lift_coming_from_above = lift_coming_from_above / n_simulations

print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ª–∏—Ñ—Ç –†–∞–π–∞–Ω–∞ –ì–æ—Å–ª–∏–Ω–≥–∞ –ø—Ä–∏–µ–¥–µ—Ç —Å–≤–µ—Ä—Ö—É: {probability_lift_coming_from_above:.3f}")
```

__–±) [5 –±–∞–ª–ª–æ–≤]__ –ì—Ä—É–ø–ø–∞ Radiohead –≤ —Å–ª—É—á–∞–π–Ω–æ–µ –≤—Ä–µ–º—è –ø–æ–¥—Ö–æ–¥–∏—Ç –∫ –ª–∏—Ñ—Ç—É –Ω–∞ 25-–æ–º —ç—Ç–∞–∂–µ –∏ –∂–¥—ë—Ç –µ–≥–æ. –û—Ü–µ–Ω–∏—Ç–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ª–∏—Ñ—Ç –ø—Ä–∏–µ–¥–µ—Ç –∫ –≥—Ä—É–ø–ø–µ Radiohead —Å–≤–µ—Ä—Ö—É. 

```python
radiohead_arrival_times = np.random.randint(0, 60, n_simulations)

lift_up_time_25 = 25
lift_down_time_25 = 55

lift_coming_from_above_radiohead = np.sum(
    (radiohead_arrival_times >= lift_down_time_25) | (radiohead_arrival_times < lift_up_time_25))

probability_lift_coming_from_above_radiohead = lift_coming_from_above_radiohead / n_simulations

print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ª–∏—Ñ—Ç –≥—Ä—É–ø–ø—ã Radiohead –ø—Ä–∏–µ–¥–µ—Ç —Å–≤–µ—Ä—Ö—É: {probability_lift_coming_from_above_radiohead:.3f}")
```

__–≤) [5 –±–∞–ª–ª–æ–≤]__  –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è –ª–∏—Ñ—Ç–∞ –¥–ª—è –†–∞–π–∞–Ω–∞ –ì–æ—Å–ª–∏–Ω–≥–∞ –∏ –≥—Ä—É–ø–ø—ã Radiohead. –ü–æ—Ö–æ–∂–∏ –ª–∏ –≤–∏–∑—É–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–µ? –ù–∞—Å–∫–æ–ª—å–∫–æ —Å–∏–ª—å–Ω–æ –æ–Ω–∏ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞?

```python
def calculate_waiting_time(arrival_times, up_time, down_time):
    waiting_times = []
    for time in arrival_times:
        if time < up_time:
            waiting_time = up_time - time
        elif up_time <= time < down_time:
            waiting_time = min(time - up_time, down_time - time)
        else:
            waiting_time = 60 - time + up_time
        waiting_times.append(waiting_time)
    return waiting_times


waiting_times_ryan = calculate_waiting_time(ryan_arrival_times, lift_up_time, lift_down_time)
waiting_times_radiohead = calculate_waiting_time(radiohead_arrival_times, lift_up_time_25, lift_down_time_25)

plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1)
plt.hist(waiting_times_ryan, bins = 30, color = 'orange', alpha = 0.7)
plt.title("–í—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è –ª–∏—Ñ—Ç–∞ –¥–ª—è –†–∞–π–∞–Ω–∞ –ì–æ—Å–ª–∏–Ω–≥–∞")
plt.xlabel("–í—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è (–º–∏–Ω—É—Ç—ã)")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤")

plt.subplot(1, 2, 2)
plt.hist(waiting_times_radiohead, bins = 30, color = 'green', alpha = 0.7)
plt.title("–í—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è –ª–∏—Ñ—Ç–∞ –¥–ª—è –≥—Ä—É–ø–ø—ã Radiohead")
plt.xlabel("–í—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è (–º–∏–Ω—É—Ç—ã)")
plt.ylabel("–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–æ–≤")

plt.tight_layout()
plt.show()
```

–ò–∑ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –≤–∏–¥–Ω–æ, —á—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –¥—Ä—É–≥ –æ—Ç –¥—Ä—É–≥–∞.

–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è –†–∞–π–∞–Ω–∞ –ì–æ—Å–ª–∏–Ω–≥–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è —Å–æ—Å—Ä–µ–¥–æ—Ç–æ—á–µ–Ω–æ –æ–∫–æ–ª–æ –º–∏–Ω–∏–º–∞–ª—å–Ω—ã—Ö –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –Ω–µ—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –¥–≤—É–º—è –ø–∏–∫–∞–º–∏. –≠—Ç–æ —Å–≤—è–∑–∞–Ω–æ —Å —Ç–µ–º, —á—Ç–æ –ª–∏—Ñ—Ç –æ—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ—Ç—Å—è –Ω–∞ –µ–≥–æ —ç—Ç–∞–∂–µ (3-–º) –≤—Å–∫–æ—Ä–µ –ø–æ—Å–ª–µ –Ω–∞—á–∞–ª–∞ —Å–≤–æ–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞ –≤–≤–µ—Ä—Ö –∏ –Ω–µ–∑–∞–¥–æ–ª–≥–æ –¥–æ –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞ –≤–Ω–∏–∑, —Å–æ–∑–¥–∞–≤–∞—è –¥–≤–∞ –ø–µ—Ä–∏–æ–¥–∞, –∫–æ–≥–¥–∞ –æ–∂–∏–¥–∞–Ω–∏–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ.

–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –¥–ª—è –≥—Ä—É–ø–ø—ã Radiohead –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –±–æ–ª–µ–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏ –æ–∂–∏–¥–∞–Ω–∏—è —Å –æ–¥–Ω–∏–º —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–º –ø–∏–∫–æ–º. –≠—Ç–æ –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏–µ–º 25-–≥–æ —ç—Ç–∞–∂–∞ –±–ª–∏–∂–µ –∫ —Å–µ—Ä–µ–¥–∏–Ω–µ –ø—É—Ç–∏ –ª–∏—Ñ—Ç–∞, –∏–∑-–∑–∞ —á–µ–≥–æ –≤—Ä–µ–º–µ–Ω–∞ –æ–∂–∏–¥–∞–Ω–∏—è –±–æ–ª–µ–µ —Å–±–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∞–Ω—ã –∏ —Å–∫–ª–æ–Ω–Ω—ã –∫ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—é –æ–¥–Ω–æ–≥–æ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–≥–æ –ø–∏–∫–∞.

–û–±–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–µ –ø–æ—Ö–æ–∂–∏ –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ. –†–∞–∑–ª–∏—á–∏—è –≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è—Ö –æ—Ç—Ä–∞–∂–∞—é—Ç —Ä–∞–∑–Ω—ã–µ —ç—Ç–∞–∂–∏, –Ω–∞ –∫–æ—Ç–æ—Ä—ã—Ö –æ–∂–∏–¥–∞—é—Ç –ª–∏—Ñ—Ç –†–∞–π–∞–Ω –ì–æ—Å–ª–∏–Ω–≥ –∏ –≥—Ä—É–ø–ø–∞ Radiohead, –∞ —Ç–∞–∫–∂–µ —Ä–∞–∑–Ω—É—é —Å—Ç—Ä—É–∫—Ç—É—Ä—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∏—Ö –æ–∂–∏–¥–∞–Ω–∏—è, –∑–∞–≤–∏—Å—è—â—É—é –æ—Ç –≤—Ä–µ–º–µ–Ω–∏ –ø—Ä–∏—Ö–æ–¥–∞ –∫ –ª–∏—Ñ—Ç—É.

__–≥) [5 –±–∞–ª–ª–æ–≤]__ –£ –°—Ç–∏–≤–∞ –†–æ–¥–∂–µ—Ä—Å–∞ –∫—É—á–∞ —Å–≤–æ–±–æ–¥–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –∏ –æ–Ω –∫–∞—Ç–∞–µ—Ç—Å—è –Ω–∞ –ª–∏—Ñ—Ç–µ –≤–µ—Å—å –¥–µ–Ω—å. –ù–∞—Ä–∏—Å—É–π—Ç–µ –¥–ª—è –°—Ç–∏–≤–∞ –¥–≤–∞ –≥—Ä–∞—Ñ–∏–∫–∞: 

- –í –ø–µ—Ä–≤–æ–º –ø–æ –æ—Å–∏ –∞–±—Å—Ü–∏—Å—Å –æ—Ç–ª–æ–∂–∏—Ç–µ —ç—Ç–∞–∂, —Å –∫–æ—Ç–æ—Ä–æ–≥–æ —Å—Ç–∞—Ä—Ç—É–µ—Ç –°—Ç–∏–≤, –∞ –ø–æ –æ—Å–∏ –æ—Ä–¥–∏–Ω–∞—Ç –æ—Ç–ª–æ–∂–∏—Ç–µ –æ—Ü–µ–Ω–∫—É –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ç–æ–≥–æ, —á—Ç–æ –ª–∏—Ñ—Ç –ø—Ä–∏–¥—ë—Ç —Å–≤–µ—Ä—Ö—É. 
- –í –ø–µ—Ä–≤–æ–º –ø–æ –æ—Å–∏ –∞–±—Å—Ü–∏—Å—Å –æ—Ç–ª–æ–∂–∏—Ç–µ —ç—Ç–∞–∂, —Å –∫–æ—Ç–æ—Ä–æ–≥–æ —Å—Ç–∞—Ä—Ç—É–µ—Ç –°—Ç–∏–≤, –∞ –ø–æ –æ—Å–∏ –æ—Ä–¥–∏–Ω–∞—Ç —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è –ª–∏—Ñ—Ç–∞.

```python
floors = np.arange(0, 31)
probability_coming_from_above = []

for floor in floors:
    lift_up_time = floor
    lift_down_time = 60 - floor
    probability = ((60 - lift_down_time) + lift_up_time) / 60
    probability_coming_from_above.append(1 - probability)

average_waiting_time = []

for floor in floors:
    if floor <= 15:
        avg_time = (floor + (60 - floor)) / 2
    else:
        avg_time = ((60 - floor) + floor) / 2
    average_waiting_time.append(avg_time)

plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1)
plt.plot(floors, probability_coming_from_above, marker = 'o', linestyle = '-', color = 'yellow')
plt.title("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø—Ä–∏–µ–∑–¥–∞ –ª–∏—Ñ—Ç–∞ —Å–≤–µ—Ä—Ö—É –¥–ª—è –°—Ç–∏–≤–∞ –†–æ–¥–∂–µ—Ä—Å–∞")
plt.xlabel("–≠—Ç–∞–∂")
plt.ylabel("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å")

plt.subplot(1, 2, 2)
plt.plot(floors, average_waiting_time, marker = 'o', linestyle = '-', color = 'green')
plt.title("–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è –ª–∏—Ñ—Ç–∞ –¥–ª—è –°—Ç–∏–≤–∞ –†–æ–¥–∂–µ—Ä—Å–∞")
plt.xlabel("–≠—Ç–∞–∂")
plt.ylabel("–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –æ–∂–∏–¥–∞–Ω–∏—è (–º–∏–Ω—É—Ç—ã)")

plt.tight_layout()
plt.show()
```

–ö–∞–∫ –¥—É–º–∞–µ—Ç–µ, –±—É–¥–µ—Ç –ª–∏ –Ω–∞–±–ª—é–¥–∞—Ç—å—Å—è —á—Ç–æ-—Ç–æ –ø–æ—Ö–æ–∂–µ–µ –≤ —Ä–µ–∞–ª—å–Ω–æ–π –∂–∏–∑–Ω–∏? –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –ø–æ–Ω–∞–±–ª—é–¥–∞—Ç—å –∑–∞ –ª–∏—Ñ—Ç–æ–º —É —Å–µ–±—è –¥–æ–º–∞ –∏ –≤ –æ—Ñ–∏—Å–µ. –ù–∞ –∫–∞–∫–æ–º —ç—Ç–∞–∂–µ –≤—ã –∂–∏–≤—ë—Ç–µ/—Ä–∞–±–æ—Ç–∞–µ—Ç–µ? –í—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –ª–∏ –¥–ª—è –≤–∞—Å [–ø–∞—Ä–∞–¥–æ–∫—Å –ª–∏—Ñ—Ç–∞?](https://en.wikipedia.org/wiki/Elevator_paradox) 

–£—á–∏—Ç—ã–≤–∞—è, —á—Ç–æ –±—ã–ª–æ –ø—Ä–æ–≤–µ–¥–µ–Ω–æ 10000 —Å–∏–º—É–ª—è—Ü–∏–π, —Ç–æ –º–æ–∂–Ω–æ –ø–æ–Ω—è—Ç—å, —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ-—Ç–æ –ø–æ—Ö–æ–∂–µ–µ –ø—Ä–æ–∏–∑–æ–π–¥—ë—Ç –≤ —Ä–µ–∞–ª—å–Ω–æ–π –∂–∏–∑–Ω–∏ ‚Äì –≤–µ–ª–∏–∫–∞ (–Ω–µ –±—É–¥–µ–º —Å—á–∏—Ç–∞—Ç—å, —á—Ç–æ –ø–µ—Ä—Å–æ–Ω–∞–∂–∏ –Ω–µ –º–æ–≥—É—Ç –≤–æ–æ–±—â–µ –ø–µ—Ä–µ—Å–µ—á—å—Å—è –ø–æ –æ–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω—ã–º –æ–±—Å—Ç–æ—è—Ç–µ–ª—å—Å—Ç–≤–∞–º)

## –ó–∞–¥–∞—á–∞ 4: –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ (30 –±–∞–ª–ª–æ–≤)

–û—Ç —Å—Ç–∞–Ω—Ü–∏–∏ –†–æ–º–∞—à–∫–æ–≤–æ –¥—Ä—É–≥ –∑–∞ –¥—Ä—É–≥–æ–º –≤ –æ–¥–Ω–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏ –æ—Ç—Ö–æ–¥–∏—Ç $100$ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–æ–≤ —Å –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–º –≤ –æ–¥–Ω—É –º–∏–Ω—É—Ç—É. –°—Ç–∞—Ä—Ç–æ–≤—ã–µ —Å–∫–æ—Ä–æ—Å—Ç–∏ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–æ–≤ -- –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ, —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω—ã–µ –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ $[0;1]$ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã. –ö–æ–≥–¥–∞ –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä—ã–π –ø–∞—Ä–æ–≤–æ–∑–∏–∫ –¥–æ–≥–æ–Ω—è–µ—Ç –±–æ–ª–µ–µ –º–µ–¥–ª–µ–Ω–Ω—ã–π, –æ–Ω —Å–Ω–∏–∂–∞–µ—Ç —Å–≤–æ—é —Å–∫–æ—Ä–æ—Å—Ç—å –¥–æ —Å–∫–æ—Ä–æ—Å—Ç–∏ –¥–æ–≥–Ω–∞–Ω–Ω–æ–≥–æ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∞.

–ß–µ—Ä–µ–∑ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –¥–ª–∏–Ω–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª –≤—Ä–µ–º–µ–Ω–∏ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ —Ä–∞–∑–æ–±—å—é—Ç—Å—è –Ω–∞ –Ω–µ—Å–∫–æ–ª—å–∫–æ –≥—Ä—É–ø–ø –∏ –¥–∞–ª–µ–µ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø –º–µ–Ω—è—Ç—å—Å—è –Ω–µ –±—É–¥–µ—Ç. –ù–∞–∑–æ–≤—ë–º —ç—Ç–æ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø —Ñ–∏–Ω–∞–ª—å–Ω—ã–º. –ü—Ä–æ–≤–µ–¥–∏—Ç–µ $1000$ —Å–∏–º—É–ª—è—Ü–∏–π –∏ –Ω–∞–π–¥–∏—Ç–µ –æ—Ç–≤–µ—Ç—ã –Ω–∞ —Å–ª–µ–¥—É—é—â–∏–µ –≤–æ–ø—Ä–æ—Å—ã:

__–∞) [5 –±–∞–ª–ª–æ–≤]__  –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ø—è—Ç—ã–π –ø–∞—Ä–æ–≤–æ–∑–∏–∫ -- –ª–∏–¥–µ—Ä –æ—á–µ—Ä–µ–¥–Ω–æ–π –≥—Ä—É–ø–ø—ã?

```python
num_trains = 100
num_simulations = 1000
```

```python
leader_counts = 0

for _ in range(num_simulations):
    speeds = np.random.uniform(0, 1, num_trains)
    leaders = [speeds[0]]

    for i in range(1, num_trains):
        if speeds[i] > leaders[-1]:
            leaders.append(speeds[i])

    if speeds[4] in leaders:
        leader_counts += 1

probability_p5_leader = leader_counts / num_simulations

print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ø—è—Ç—ã–π –ø–∞—Ä–æ–≤–æ–∑–∏–∫ ‚Äì –ª–∏–¥–µ—Ä: {probability_p5_leader}")
```

__–±) [5 –±–∞–ª–ª–æ–≤]__ –ù–∞ —Å–∫–æ–ª—å–∫–æ —Ñ–∏–Ω–∞–ª—å–Ω—ã—Ö –≥—Ä—É–ø–ø –≤ —Å—Ä–µ–¥–Ω–µ–º —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏?

```python
group_counts = []

for _ in range(num_simulations):
    speeds = np.random.uniform(0, 1, num_trains)
    leaders = [speeds[0]]

    for i in range(1, num_trains):
        if speeds[i] > leaders[-1]:
            leaders.append(speeds[i])

    group_counts.append(len(leaders))

average_final_groups = np.mean(group_counts)
print(f"–í —Å—Ä–µ–¥–Ω–µ–º —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –Ω–∞ {round(average_final_groups, 0):.0f} –≥—Ä—É–ø–ø")
```

__–≤) [10 –±–∞–ª–ª–æ–≤]__ –ù–∞—Ä–∏—Å—É–π—Ç–µ —É—Å—Ä–µ–¥–Ω—ë–Ω–Ω–æ–µ –ø–æ —Å–∏–º—É–ª—è—Ü–∏—è–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥—Ä—É–ø–ø –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –≤—Ä–µ–º–µ–Ω–∏ –æ—Ç —Å—Ç–∞—Ä—Ç–∞. –ü–æ –≥–æ—Ä–∏–∑–æ–Ω–∞—Ç–ª—å–Ω–æ–π –æ—Å–∏ ‚Äî –≤—Ä–µ–º—è, –ø—Ä–æ—à–µ–¥—à–µ–µ –æ—Ç —Å—Ç–∞—Ä—Ç–∞. –ü–æ –≤–µ—Ä—Ç–∏–∫–∞–ª—å–Ω–æ–π –æ—Å–∏ ‚Äî —Å—Ä–µ–¥–Ω–µ–µ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø. 

```python
average_group_counts = np.zeros(num_trains)

for _ in range(num_simulations):
    speeds = np.random.uniform(0, 1, num_trains)
    times = [0]

    for i in range(1, num_trains):
        time_to_catch_up = np.inf
        for j in range(i):
            if speeds[i] > speeds[j]:
                catch_up_time = j + 1 + (1 / (speeds[i] - speeds[j]))
                if catch_up_time < time_to_catch_up:
                    time_to_catch_up = catch_up_time
        times.append(time_to_catch_up)

    for time_step in range(num_trains):
        group_count = len(set([time for time in times if time <= time_step + 1]))
        average_group_counts[time_step] += group_count

average_group_counts /= num_simulations

time_steps = np.arange(1, len(average_group_counts) + 1)

plt.figure(figsize=(10, 6))
plt.plot(time_steps, average_group_counts, label='–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥—Ä—É–ø–ø')
plt.xlabel('–®–∞–≥–∏ –≤—Ä–µ–º–µ–Ω–∏ –æ—Ç –Ω–∞—á–∞–ª–∞')
plt.ylabel('–°—Ä–µ–¥–Ω–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥—Ä—É–ø–ø')
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è –≥—Ä—É–ø–ø —Å–æ –≤—Ä–µ–º–µ–Ω–µ–º')
plt.legend()
plt.grid(True)
plt.show()
```

__–≥) [10 –±–∞–ª–ª–æ–≤]__  –û—Ü–µ–Ω–∏—Ç–µ —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è, –≤ —Ç–µ—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –Ω–∞ —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø. –û—Ü–µ–Ω–∏—Ç–µ –¥–ª—è —ç—Ç–æ–π –∂–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –¥–∏—Å–ø–µ—Ä—Å–∏—é.

```python
final_group_formation_times = []

for _ in range(num_simulations):
    speeds = np.random.uniform(0, 1, num_trains)
    formation_time = 0

    for i in range(1, num_trains):
        for j in range(i):
            if speeds[i] > speeds[j]:
                current_formation_time = j + 1 + (1 / (speeds[i] - speeds[j]))
                formation_time = max(formation_time, current_formation_time)

    final_group_formation_times.append(formation_time)

average_formation_time = np.mean(final_group_formation_times)
variance_formation_time = np.var(final_group_formation_times)

print(f"–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è, –≤ —Ç–µ—á–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –ø–∞—Ä–æ–≤–æ–∑–∏–∫–∏ —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –Ω–∞ —Ñ–∏–Ω–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ –≥—Ä—É–ø–ø: {average_formation_time}")
print(f"–í—Ä–µ–º—è –¥–∏—Å–ø–µ—Ä—Å–∏–∏ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏—è —Ñ–∏–Ω–∞–ª—å–Ω–æ–≥–æ —á–∏—Å–ª–∞ –≥—Ä—É–ø–ø: {variance_formation_time}")
```
